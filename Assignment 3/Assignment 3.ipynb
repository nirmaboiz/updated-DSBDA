```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('airquality.csv')
```

```python
df.head()
```

```python
df.columns
```

# Pie Chart

```python
data=[df["Ozone"].mean(),df["Wind"].mean(),df["Temp"].mean(),df["Solar.R"].mean()]
colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode=(0.1, 0.1, 0.1, 0.1)
labels_name=["Ozone","Wind","Temp","Solar.R"]
plt.pie(data,explode=explode,labels = labels_name, autopct='%1.1f%%', colors=colors,shadow=True, radius=5)
```

# Bar Chart 

```python
plt.show()
```

```python
h = df.iloc[1:21,4]
y_pos = np.arange(len(h))
v = range(1,21)
plt.figure(figsize=(15,8))
plt.bar(y_pos,h,align = 'center', alpha = 0.5)
plt.xticks(y_pos,v)
plt.ylabel('range') 
plt.xlabel("Days")
plt.title('Temperature for 1st-20th May')
plt.show()
```

# Box Plot

```python
plt.rcParams['figure.autolayout'] = True
df.boxplot(column=['Ozone','Wind','Temp','Solar.R'])
plt.xlabel('Air Quality')
plt.ylabel('Values')
plt.show()
```

# Histogram

```python
h = df.iloc[:,-1]
plt.hist(h, bins='auto')
plt.title('Histogram of Solar Radiation')
plt.ylabel("Frequency")
plt.xlabel("Solar Radiation")
```

# Line Graph

```python
plt.plot(df['Ozone'], df['Wind'], color='red', marker='o')
plt.title('Ozone Vs Wind', fontsize=14)
plt.xlabel('Wind', fontsize=14)
plt.ylabel('Ozone', fontsize=14)
plt.grid(True)
plt.show()
```

# Scatterplot 

```python
sns.scatterplot(data=df, x="Ozone", y="Wind")
```

# Scatter Matrices

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, alpha = 0.2, figsize = (6, 6))
```

# Pairplot

```python
sns.set(style = "ticks")
sns.pairplot(df) 
plt.show()
```
